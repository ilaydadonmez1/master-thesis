```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Load the CSV file into a DataFrame
file_path = '/Users/ilaydadonmez/Desktop/InitialAnalysis.csv'
df = pd.read_csv(file_path, delimiter=';')

# Create output directory if it doesn't exist
output_dir = '/Users/ilaydadonmez/Desktop/Initial Analysis'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Summarize the DataFrame
#summary = df.describe(include='all')
#info = df.info()

# Display the summary
#print(summary)
#print(info)
```

```python
# Plot a histogram of Modularity
plt.figure(figsize=(10, 6))
plt.hist(df['Mod_2'], bins=10, edgecolor='black')
plt.xlabel('Modularity')
plt.ylabel('Frequency')
plt.title('Distribution of Modularity')
plt.grid(True)
histogram_path = os.path.join(output_dir, 'Distribution_of_Modularity.png')
plt.savefig(histogram_path)
plt.show()

# Boxplot of Modularity by Industry
plt.figure(figsize=(12, 6))
sns.boxplot(x='Ind', y='Mod_2', data=df)
plt.xticks(rotation=90)
plt.xlabel('Industry')
plt.ylabel('Modularity')
plt.title('Modularity by Industry')
boxplot_path = os.path.join(output_dir, 'Modularity_by_Industry.png')
plt.savefig(boxplot_path)
plt.show()

# Scatter Plot of Size vs Modularity
plt.figure(figsize=(10, 6))
plt.scatter(df['Size'], df['Mod_2'], c='blue', alpha=0.5)
plt.xlabel('Size')
plt.ylabel('Modularity')
plt.title('Size vs Modularity')
plt.grid(True)
scatter_path = os.path.join(output_dir, 'Size_vs_Modularity.png')
plt.savefig(scatter_path)
plt.show()

# Pairplot of selected metrics
selected_columns = ['Size', 'Mod_2', 'Edges', 'Mutual', 'Cycle3', 'Triangle']
sns.pairplot(df[selected_columns])
plt.suptitle('Pairplot of Selected Metrics', y=1.02)
pairplot_path = os.path.join(output_dir, 'Pairplot_of_Selected_Metrics.png')
plt.savefig(pairplot_path)
plt.show()

# Correlation heatmap of numeric variables
plt.figure(figsize=(14, 10))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Variables')
heatmap_path = os.path.join(output_dir, 'Correlation_Heatmap.png')
plt.savefig(heatmap_path)
plt.show()
```
